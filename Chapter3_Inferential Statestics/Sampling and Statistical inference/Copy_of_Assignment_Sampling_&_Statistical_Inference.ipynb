# <u>Problem 3

Screenshot 2020-11-30 at 5.28.32 PM.png

```
import scipy.stats as stats
mu = 100
sigma = 25
n = 5

# SEM
sem = sigma / (n ** 0.5)

# Z score between 80 to 120
z_score_80 = (80 - mu) / sem

z_score_120 = (120 - mu) / sem

# CDF
probability_between_80_120 = stats.norm.cdf(z_score_120) - stats.norm.cdf(z_score_80)

# Calculate the degrees of freedom for chi-squared distribution
df = n - 1

# Calculate the probability that S exceeds 41.7 using the chi-squared distribution

probability_s_exceeds_41_7 = 1 - stats.chi2.cdf(41.7**2 * df / sigma**2, df)



print("Probability that xbar is between 80 and 120:", probability_between_80_120)

print("Probability that S exceeds 41.7:", probability_s_exceeds_41_7)
```

```
# Above question follows t distribution therefore
import scipy.stats as stats

n = 5
df = n - 1
value = 1.533

# CDF
probability_exceeds_1_533 = 1 - stats.t.cdf(value, df)

print("Probability that (xbar - 100) / (S / 5) exceeds 1.533:", probability_exceeds_1_533)
```
