# <b><u> Problem 1 </u> </b>

## <b>The average IQ of a sample of 50 university students was found to be 105. Carry out a statistical test to determine whether the average IQ of university students is greater than 100, assuming that IQs are normally distributed. It is known from previous studies that the standard deviation of IQs among students is approximately 20. </b>

```
sample_mean = 105
population_mean = 100
sample_size = 50
population_standard_deviation = 20

# Perform z-test
z_score = (sample_mean - population_mean) / (population_standard_deviation / (sample_size ** 0.5))
print(z_score)

# Using a standard normal distribution to calculate p-value
import scipy.stats as stats
p_value = 1 - stats.norm.cdf(z_score)                        # p-value tells the strength of the alternative hypothesis to reject null hypothesis
print(p_value)
alpha = 0.05    # alpha is 5% it will indicates the level of significance means if we reject null hypothesis even if its true then in that case
                # we need to specify level of significance.

if p_value < alpha:

    print("Reject the null hypothesis. There is enough evidence to suggest that the average IQ of university students is greater than 100.")

else:

    print("Fail to reject the null hypothesis. There is not enough evidence to suggest that the average IQ of university students is greater than 100.")
```

# <b><u> Problem 2 </u> </b>

## <b>Carry out a statistical test to assess whether the standard deviation of the heights of 10- year-old children is equal to 3cm, based on the random sample of 5 heights in cm given below. Assume that heights are normally distributed.

## 124, 122, 130, 125, 132 </b>

```
import numpy as np

import scipy.stats as stats

sample_heights = [124, 122, 130, 125, 132]

population_variance = 3

sample_variance = np.var(sample_heights, ddof=1)  # ddof=1 for unbiased sample variance


df = len(sample_heights) - 1

test_statistic = df * sample_variance / population_variance

alpha = 0.05
alpha_two_tailed = alpha / 2

critical_value = stats.chi2.ppf(1 - alpha_two_tailed, df)

if test_statistic > critical_value:
    print("Reject the null hypothesis. The population standard deviation is not equal to 3 cm.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to suggest that the population standard deviation is different from 3 cm.")

print("Test Statistic:", test_statistic)
print("Critical Value:", critical_value)
```

# <b><u> Problem 3 </u> </b>

## <b>In a one-year mortality investigation, 45 of the 250 ninety-year-olds present at the start of the investigation died before the end of the year. Assuming that the number of deaths has a Bin(250,q) distribution, test whether this result is consistent with a mortality rate of q = 0.2 for this age. </b>

```
import scipy.stats as stats

observed_deaths = 45
total_individuals = 250
expected_mortality_rate = 0.2

expected_deaths = total_individuals * expected_mortality_rate

z_score = (observed_deaths - expected_deaths) / (total_individuals * expected_mortality_rate * (1 - expected_mortality_rate))**0.5


p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))


alpha = 0.05

if p_value < alpha:

    print("Reject the null hypothesis. The observed mortality result is inconsistent with a mortality rate of q = 0.2.")

else:

    print("Fail to reject the null hypothesis. The observed mortality result is consistent with a mortality rate of q = 0.2.")



print("Z-Score:", z_score)

print("P-Value:", p_value)
```

# <b><u> Problem 4 </u> </b>

## <b>A new gene has been identified that makes carriers particularly susceptible to a particular degenerative disease. In a random sample of 250 adult males born in the UK, 8 were found to be carriers of the disease. Test whether the proportion of adult males born in the UK carrying the gene is less than 10%. </b>



```
import scipy.stats as stats

sample_size = 250
observed_carriers = 8
hypothesized_proportion = 0.10


expected_carriers = sample_size * hypothesized_proportion

z_score = (observed_carriers - expected_carriers) / (sample_size * hypothesized_proportion * (1 - hypothesized_proportion))**0.5

p_value = stats.norm.cdf(z_score)

alpha = 0.05


if p_value < alpha:

    print("Reject the null hypothesis. The proportion of adult males carrying the gene is less than 10%.")

else:

    print("Fail to reject the null hypothesis. There is not enough evidence to suggest that the proportion of adult males carrying the gene is less than 10.")



print("Z-Score:", z_score)

print("P-Value:", p_value)
```
